```python
df = sns.load_dataset('titanic')
df
```

```python
# Plot a count-plot. Counting something based on a category. (So, when you have discrete categorisation of a set)
sns.countplot(x = 'sex',hue = 'who',data = df)
# By using 'hue', you can sub-categorize and represent by color.
```

```python
# Kernel density estimate - plot the distribution of a dataset.
# We assume that the dataset is discrete, that we can produce some sort of histogram - data vs. count. 
# Then, we can produce - data vs. density.
sns.kdeplot(data = df['age'])
```

```python
# Similarly, we can plot KDE graph AND histogram.
penguins = sns.load_dataset("penguins")
sns.distplot(data=penguins, x="flipper_length_mm")
```

```python
sns.scatterplot(x = 'age', y = 'fare', data = df, hue = 'who')
```

```python
df = sns.load_dataset('iris')
sns.scatterplot(x = 'sepal_length', y= 'petal_length', data = df, hue = 'species')
```

```python
sns.jointplot(x = 'sepal_length', y= 'petal_length', data = df, kind = 'kde')
```

```python
sns.jointplot(x = 'sepal_length', y= 'petal_length', data = df, kind = 'reg')
```

```python
# Plot multiple scatterplots
sns.pairplot(df)
```

```python
corr = df.corr() # Compute pairwise correlation of a dataframe's data (columns)
sns.heatmap(corr,cmap = ['red', 'green', 'blue'], vmin = -0.5, vmax = 0.6, annot = True)
```
